```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as MSE
```

```python
arr = np.array(pd.read_csv("astro_ps.csv"))
np.isnan(arr)
```

```python
plt.figure(figsize=(12,12))
plt.plot(arr[:,0],arr[:,1])
plt.xlabel('Val1')
plt.ylabel('Time')
```

```python
indhigh,_ = find_peaks(arr[:,1])
```

```python
plt.plot(arr[indhigh,0],arr[indhigh,1])
```

```python
f_inst = []
for i in range(indhigh.shape[0]-1):
    f_inst.append(1/(arr[indhigh[i+1],0]-arr[indhigh[i],0]))
```

```python
f_inst
```

```python
plt.plot(arr[indhigh[1:],0],f_inst)
```

```python
X= arr[indhigh[1:],0].reshape(-1,1)
y = np.array(f_inst)
y.shape
```

```python
Z=np.squeeze(np.array([X**(-1),X**(-2)])).T
Z.shape
```

```python
model = LinearRegression()
model = model.fit(Z,y)
```

```python
params = model.coef_
params
```

```python
model.score(Z,y)
```

```python
der_params = np.array([-params[0],-2*params[1]])
der_params
```

```python
ti= np.squeeze(np.array([arr[:-1,0]**(-1),arr[:-1,0]**(-2)]))
ti
```

```python
f_fit = (np.matmul(params,ti))
```

```python
f_der_fit = np.matmul(der_params,np.squeeze(np.array([arr[:-1,0]**(-2),arr[:-1,0]**(-3)])))
```

```python
K = (np.power(f_fit[0:-7],-11/3)*(f_der_fit[0:-7]))
K
```

```python
plt.plot(np.power(np.abs(K),3/5))
```
